```r
library('tidyverse')
```

```r
wind_data <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
```

Each row in the table represents a single turbine.

Make a bar chart 📊 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_data |> head()
```

```r
data = wind_data |> 
    mutate(prov2 = province_territory |> fct_infreq() |> fct_rev())
```

```r
options(repr.plot.res = 250, repr.plot.height = 4, repr.plot.width = 6)

p = ggplot(data, aes(y = prov2)) + geom_bar() +
    labs(x="Turbine Count", y= "Province / Territory")

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
data3 = wind_data |> 
    mutate(manufactur_sort = fct_lump(manufacturer, n=6) |> 
        fct_infreq() |> 
        fct_relevel('Other', after=Inf) |>
        fct_rev())
```

```r
options(repr.plot.res = 250, repr.plot.height = 4, repr.plot.width = 6)

p = ggplot(data3, aes(y = manufactur_sort)) + geom_bar() + 
    labs(x= 'Turbine Count', y = 'Manufacturer')

p
```
